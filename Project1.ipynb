```python
# create a file list that will be used for appending
file_list=['http://web.mta.info/developers/data/nyct/turnstile/turnstile_220305.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220312.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220319.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220326.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220402.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220409.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220416.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220423.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220430.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220507.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220514.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220521.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220528.txt','http://web.mta.info/developers/data/nyct/turnstile/turnstile_220604.txt']
file_list
```

```python
 # create a for loop to append weekly data files
main_df = pd.DataFrame(pd.read_csv(file_list[0]))
  
for i in range(1,len(file_list)):
    data = pd.read_csv(file_list[i])
    weekly_df= pd.DataFrame(data)
    main_df = pd.concat([main_df,weekly_df], ignore_index = True)
print(main_df)
```

```python
min_max_df=main_df.groupby(['station', 'station_id', 'control_area','device_position','line','division','date']).agg(hour_min=('hour','min'),hour_max=('hour','max'), entry_min=('entry','min'), entry_max=('entry', 'max'), exit_min=('exit', 'min'), exit_max=('exit','max')).reset_index()

```

```python
# rename columns and keep the changes
merged_df['min_hour_key'] =(merged_df.hour == merged_df.hour_min)
merged_df['max_hour_key'] =(merged_df.hour == merged_df.hour_max)
merged_df['min_entry_key'] =(merged_df.entry == merged_df.entry_min)
merged_df['max_entry_key'] =(merged_df.entry == merged_df.entry_max)
merged_df['min_exit_key'] =(merged_df.exit == merged_df.exit_min)
merged_df['max_exit_key'] =(merged_df.exit == merged_df.exit_max)
```

```python
merged_df.head(20)
```

```python
merged_df['pls_min_exit'] =(merged_df.min_hour_key == merged_df.min_exit_key)
merged_df['pls_max_exit'] =(merged_df.max_hour_key == merged_df.max_exit_key)
merged_df['pls_min_entry'] =(merged_df.min_hour_key == merged_df.min_entry_key)
merged_df['pls_max_entry'] =(merged_df.max_hour_key == merged_df.max_entry_key)                             
```

```python
cross_check_df=merged_df.groupby(['station', 'station_id', 'control_area','device_position','line','division','date']).agg(plausible_min_exit=('pls_min_exit','min'),plausible_max_exit=('pls_max_exit','min'),plausible_min_entry=('pls_min_entry','min'),plausible_max_entry=('pls_max_entry','min')).reset_index()

```

```python
cross_check_df.head(20)
```

```python
merged_df=pd.merge(merged_df,cross_check_df, on=['station', 'station_id', 'control_area','device_position','line','division','date'], how='left')
```

```python
merged_df.head(40)
```

```python
merged_df['ultimate_crosscheck'] = ((merged_df.plausible_min_exit== True) &  (merged_df.plausible_max_exit == True) & (merged_df.plausible_min_entry == True) & (merged_df.plausible_max_entry == True))


```

```python
daily_merged = (merged_df.min_hour_key == True) | (merged_df.max_hour_key == True)
```

```python
daily_merged_df=merged_df[daily_merged]
```

```python
daily_merged_df.sort_values(['station', 'station_id', 'control_area','device_position','line','division','date'])
```

```python
daily_merged_df.info()
```

```python
cleared_master_df=daily_merged_df.groupby(['station', 'station_id', 'control_area','device_position','line','division','date']).agg(time_min=('time','min'),time_max=('time','max'), hour_min=('hour','min'),hour_max=('hour','max'), entry_min=('entry','min'), entry_max=('entry', 'max'), exit_min=('exit', 'min'), exit_max=('exit','max'), cross_check_mean=('ultimate_crosscheck','mean'),cross_check_min=('ultimate_crosscheck','min')).reset_index()
cleared_master_df.tail(40)


```

```python
cleared_master_df['daily_entry'] = cleared_master_df.entry_max- cleared_master_df.entry_min
cleared_master_df['daily_exit'] = cleared_master_df.exit_max- cleared_master_df.exit_min
cleared_master_df['hour_interval'] = cleared_master_df.hour_max - cleared_master_df.hour_min
cleared_master_df['avg_hourly_entry'] = cleared_master_df.daily_entry / cleared_master_df.hour_interval
cleared_master_df['avg_hourly_exit'] = cleared_master_df.daily_exit / cleared_master_df.hour_interval
```

```python
cleared_master_df.head(40)
```

```python
cleared_master_df['limit_in_hour']=1800
cleared_master_df['over_hourly_exit'] = cleared_master_df.avg_hourly_exit > cleared_master_df.limit_in_hour 
cleared_master_df['over_hourly_entry'] = cleared_master_df.avg_hourly_entry > cleared_master_df.limit_in_hour 
```
